### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4,1],[2,3]],dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lambda_val = eigenvalues[i]
    lhs = A @ v
    rhs = lambda_val * v
    assert np.allclose(lhs, rhs), f"Verification failed for eigenpair {i}"
    print(f"Verified for eigenpair {i}: A @ v = {lhs}, λ * v = {rhs}")

# Plot
fig, ax = plt.subplots()
ax.quiver(0, 0, eigenvectors[0, 0] * eigenvalues[0], eigenvectors[1, 0] * eigenvalues[0], angles='xy', scale_units='xy', scale=1, color='r', label='Eigenvector 1')
ax.quiver(0, 0, eigenvectors[0, 1] * eigenvalues[1], eigenvectors[1, 1] * eigenvalues[1], angles='xy', scale_units='xy', scale=1, color='b', label='Eigenvector 2')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8)
ax.grid(); ax.legend()
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

#### Your Geometric Explanation(4.3):

When we apply a **matrix** to a vector, it usually changes both the direction and the length of the vector. It is like pushing and pulling on the vector, stretching or rotating it.

However, there are special vectors called **eigenvectors**. When a matrix is applied to an eigenvector, the eigenvector **does not change its direction**. It might get longer or shorter, or even flip to the opposite direction, but it will always stay along the same line it was on before.

An **eigenvalue** is simply a number that tells us **how much** an eigenvector was stretched or squeezed by the matrix. For example, if an eigenvalue is 2, the eigenvector becomes twice as long. If it's 0.5, it becomes half as long. If it's -1, it flips to the exact opposite direction but keeps the same length.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt=np.linalg.svd(X,full_matrices=False)# YOUR CODE HERE

X_reconstructed =U @ np.diag(S) @ Vt# YOUR CODE HERE — reconstruct from U, S, Vt
X_approx=S[0] * np.outer(U[:, 0],Vt[0, :])# YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

In SVD, we decompose a matrix `X` into three components: `U`, `S`, and `Vt`.

*   **U (left singular vectors):** This matrix contains the **principal components** as its columns. These vectors represent the directions of maximum variance in the data.
*   **S (singular values):** This is a diagonal matrix (or vector) that contains the singular values in decreasing order. These values indicate the strength or importance of each principal component. Larger singular values mean more variance captured by the corresponding principal component.
*   **Vt (right singular vectors):** This matrix (transpose of V) can be thought of as how the original features are combined to form the principal components, or how data points project onto the principal components.

The matrix that contains the **directions of maximum variance** is **U** (specifically, its columns).

### Relationship Between PCA and SVD

PCA aims to find directions of maximum variance in the data. Instead of computing eigenvalues and eigenvectors of the covariance matrix, SVD directly provides this information:

- **Principal Components (scores)** → obtained from U S
- **Principal Axes (feature directions)** → obtained from V
- **Variance Explained** → determined by the squared singular values in S

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm') # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm') # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts) # YOUR CODE HERE-adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts) # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff) # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    min_val = min(np.min(expected), np.min(actual))
    max_val = max(np.max(expected), np.max(actual))
    bins = np.linspace(min_val, max_val, bins + 1)

    expected_counts, _ = np.histogram(expected, bins=bins)
    actual_counts, _ = np.histogram(actual, bins=bins)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    epsilon = 1e-10
    expected_pct = np.maximum(expected_pct, epsilon)
    actual_pct = np.maximum(actual_pct, epsilon)

    psi_value = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi_value

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)
drift_dist = np.random.normal(65, 12, 1000)

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, alpha=0.5, label='Training Distribution', density=True, color='blue')
ax.hist(drift_dist, bins=30, alpha=0.5, label='Production Distribution', density=True, color='red')
ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Occurs when the relationship between the input variables and the target variable changes over time. For example, if a model predicts house prices, and suddenly a new economic policy changes how buyers value houses, that's concept drift.
- **Covariate drift:** Occurs when the distribution of the input variables (covariates) changes over time, but the relationship between inputs and the target remains the same. For example, if a model predicts loan defaults, and the income distribution of new applicants changes (e.g., more high-income applicants), that's covariate drift.
- **PSI < 0.1:** Indicates no significant shift in the distribution; the model is likely stable.
- **PSI 0.1–0.2:** Suggests a minor shift in the distribution; worth monitoring, but often no immediate action is needed.
- **PSI > 0.2:** Indicates a major shift in the distribution; this is a strong signal that the model's performance might be degrading and requires immediate investigation or retraining.
- **Retraining trigger example:** A PSI value consistently above 0.2 for a critical input feature could trigger an automated retraining pipeline for the model.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham) # YOUR CODE HERE

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    p_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * p_ham)
    if p_word == 0:
        return 0 # Avoid division by zero if word has zero probability
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The initial probability that a message is spam, before seeing any specific word. |
| Likelihood | P(word \| Spam) | The probability of seeing the word "free" given that the message is spam. |
| Evidence | P(word) | The overall probability of seeing the word "free" in any message (spam or not spam). |
| Posterior | P(Spam \| word) | The updated probability that a message is spam, given that it contains the word "free". |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, size=sample_size, replace=False).mean() for _ in range(n_samples)]) # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size) # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Sample Means Histogram')
x_vals = np.linspace(min(sample_means), max(sample_means), 100)
ax.plot(x_vals, stats.norm.pdf(x_vals, loc=pop_mean, scale=clt_std), color='red', linestyle='--', label='Theoretical Normal PDF')
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.legend()
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Density')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*



The Central Limit Theorem (CLT) is important because it tells us that if we take many samples from a population, the distribution of their means will become approximately normal when the sample size is large enough. This happens even if the original population data is not normally distributed.

This result is very useful in statistics because many classical statistical tests, such as t-tests and confidence intervals, assume that the data follows a normal distribution. In machine learning, CLT helps in evaluating model performance, estimating parameters, and making reliable predictions based on sample data. It makes statistical analysis easier and more accurate when working with large datasets.
